### Let's see what happens when data is split instead. Let's check the MAE

```python
from sklearn.model_selection import train_test_split

train_X, val_X, train_y, val_y = train_test_split(X, y, random_state =0)

# Define the model
california_model = DecisionTreeRegressor(random_state=1)

# fit the model
california_model.fit(train_X, train_y)


# get predicted prices on validation
val_predictions = california_model.predict(val_X)
print(mean_absolute_error(val_y, val_predictions))
```
